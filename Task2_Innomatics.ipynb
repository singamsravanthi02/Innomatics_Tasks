# **Problem 1: Unique Words in a Sentence**

```python
sentence = input("Enter sentence: ")

words = sentence.split()
unique_words = set(words)

print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

# **Problem 2: Highest Salary from Employee Data**

```python
employees = {}

n = int(input("Enter number of employees: "))

for i in range(n):
    name = input("Enter employee name: ")
    salary = int(input("Enter salary: "))
    employees[name] = salary

highest_employee = max(employees, key=employees.get)

print("Highest Salary:", highest_employee, "-", employees[highest_employee])
```

# **Problem 3: Find Maximum and Minimum Values**

```python
numbers = list(map(int, input("Enter numbers separated by comma: ").split(',')))

maximum = numbers[0]
minimum = numbers[0]

for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

# **Problem 4: Count Products Above a Price Threshold**

```python
prices = list(map(int, input("Enter product prices separated by comma: ").split(',')))

count = 0
for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

# **Problem 5: Calculate Attendance Percentage**

```python
attendance = input("Enter attendance (P/A) separated by space: ").split()

present_count = attendance.count("P")
total_days = len(attendance)

percentage = (present_count / total_days) * 100

print("Attendance Percentage:", percentage)
```

# **Problem 6: Remove Duplicate Phone Numbers**

```python
phone_numbers = list(map(int, input("Enter phone numbers separated by space: ").split()))

unique_numbers = set(phone_numbers)

print("Unique phone numbers:", unique_numbers)
```

# **Problem 7: Count Character Frequency**

```python
text = input("Enter a string: ")

char_count = {}

for ch in text:
    if ch in char_count:
        char_count[ch] += 1
    else:
        char_count[ch] = 1

print(char_count)
```

# **Problem 9: Check if a Key Exists in a Dictionary**

```python
employees = {}

n = int(input("Enter number of employees: "))

for i in range(n):
    name = input("Enter employee name: ")
    employees[name] = True

search_key = input("Enter employee name to search: ")

if search_key in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

# **Problem 10: Calculate Average Marks**

```python
marks = list(map(int, input("Enter marks separated by space: ").split()))

total = sum(marks)
average = total / len(marks)

print("Average Marks:", average)
```
